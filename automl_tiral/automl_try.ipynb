```python
from autogluon.tabular import *
import pandas as pd
```

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
```

```python
train_data.head()
```

```python
train_data = train_data.drop(['Unnamed: 0','id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
train_df, val_df = train_test_split(train_data, test_size=0.2, random_state=42) 
```

```python
predictor_2=TabularPredictor(eval_metric='rmsle',path='second_try',label='Calories',verbosity=2,problem_type='regression')
```

```python
predictor_2.fit(train_df,presets='best')
```

```python
predictor_2.leaderboard()
```

```python
predictor_2.evaluate(val_df)
```

```python
my_predict=predictor_2.predict(test_data)
```

```python
my_submission=pd.concat([test_data['id'],my_predict],axis=1)
```

```python
my_submission.to_csv('first_submission.csv',index=False)
```
