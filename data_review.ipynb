```python
import pandas as pd

def clean_data(df):
    # 将 "female" 的所有实例替换为列: 'Sex' 中的 "1"
    df.loc[df['Sex'].str.lower() == "female".lower(), 'Sex'] = "1"
    # 将 "male" 的所有实例替换为列: 'Sex' 中的 "0"
    df.loc[df['Sex'].str.lower() == "male".lower(), 'Sex'] = "0"
    return df
```

```python
df=clean_data(test_data)
```

```python
df.to_csv('test.csv')
```
